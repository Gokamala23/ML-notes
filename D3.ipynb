```python
#matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
a=np.linspace(0,10,50)
b=np.sin(a)
c=np.cos(a)
```

```python
print(b)
```

```python
print(c)
```

```python
plt.plot(a,b)
plt.show()
```

```python
plt.plot(a,c)
plt.show()
```

```python
#adding titles,x,y labels
plt.plot(a,b)
plt.title('sine wave')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```

```python
#parabola
x=np.linspace(-10,10,15)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'b*')
```

```python
x=np.linspace(-2,2,20)
plt.plot(x,np.sin(x),'g--')
plt.plot(x,np.cos(x),'r+')
plt.show()
```

```python
#bar plot
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
lang=['hindi','urdu','tamil']
people=[20,22,23]
ax.bar(lang,people)
plt.xlabel('languages')
plt.ylabel('no of people')
plt.show()
```

```python
#pie chart
fig1=plt.figure()
ax=fig1.add_axes([0,0,1,1])
lang=['hindi','urdu','tamil']
people=[20,22,23]
ax.pie(people,labels=lang,autopct='%1.1f%%')
plt.show()
```

```python
#scatter plot
a=np.linspace(0,10,50)
b=np.sin(a)
c=np.cos(a)
fig2=plt.figure()
ax=fig2.add_axes([0,0,1,1])
ax.scatter(a,b)
ax.scatter(a,c)
plt.show()
```

```python
#3d scatter plot
fig3=plt.figure()
ax=fig3.add_subplot(projection='3d')
a=20*np.random.random(100)
b=np.sin(a)
c=np.cos(a)
ax.scatter(a,b,c)
plt.show()
```

```python
#seaborn
import seaborn as sns
#iris dataset
iris=sns.load_dataset('iris')
iris.head()
```

```python
sns.scatterplot(x='sepal_length',y='sepal_width',hue='species',data=iris)
plt.show()
```

```python
sns.countplot(x='petal_width', data=iris)
plt.show()
```

```python
sns.distplot(iris['petal_width'])
plt.show()
```

```python
#heat map
plt.figure(figsize=(5,5))
correlation = iris.corr(numeric_only=True)
sns.heatmap(correlation,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Greens')
```
